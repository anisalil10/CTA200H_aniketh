```python
def f(x):
    """
    This function returns the output of f(x) = x^3 - x^2 - 1
    
    Parameters:
    x : float or int
        Input value for function
        
    Returns: 
    f_x : int or array
        the quantity f(x)
    """
    f_x = x**3 - x**2 - 1
    return f_x

def df(x):
    """
    This function returns the derivative of f(x) = x^3 - x^2 - 1
    df/dx = 3x^2 - 2*x
    
    Parameters: 
    x : float or int
        Input for derivative function
    
    Returns: 
    df: float or int 
        the quantity df(x)
    """
    df = 3*(x**2) - 2*x
    return df

def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    This functions returns the Newton root of function f, derivative df, and input x.
    
    Parameters: 
    f : function
        Arbitrary mathematical function input
    df : function
        Derivative function of function x
    x0 : int or float
        Starting value of Newton iteration
    epsilon : int or float
        Minimum range around 0 for f(x) to reach
    max_iter : int or float
        Maximum iteration amount
    
    Returns: 
    x : integer or array
        the result of the Newton iterations after max_ter or below threshold epsilon value

    """
    i = 0
    x = x0
    while abs(f(x)) > epsilon:
        if df(x) == 0:
            print("Zero Division Error")
            return None
        x -= f(x)/df(x)
        if i == max_iter:
            print("Iteration failed")
            return None
        i += 1
    print(f"Found root in {i} iterations")
    return x
```

```python
print(newton(f, df, 2))
print(newton(f, df, -4))

print(newton(f, df, 10, epsilon=1e-8))
```
